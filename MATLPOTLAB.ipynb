#Other 2D Grapies

```python
n=np.array([0,1,2,3,4,5])
fig,Axis=plt.subplots(1,4,figsize=(16,5))
Axis[0].set_title("Scatter plot")
Axis[0].scatter(x, x+0.25*np.random.randn(len(x)))

Axis[1].set_title("Step plot")
Axis[1].step(n,n**2,lw=5)

Axis[2].set_title("Bar plot")
Axis[2].bar(n,n*3,align="center",width=0.5,bottom=0,color="gray",edgecolor="red")

Axis[3].set_title("fill between")
Axis[3].fill_between(x,x**2,x**3, color="green",gid='str')
```

```python
"""Draw a radar chart"""
fig=plt.figure(figsize=(6,6))
ax=fig.add_axes([0.1,0,1,1], polar=True)
t=np.linspace(0,2*np.pi,100)
ax.plot(t,t*.5,color="Black",lw=3)
```

```python
n=np.random.randn(10000)
fix, ax=plt.subplots(1,2,figsize=(16,6))
ax[0].set_title('Histogram')
ax[0].hist(n,color='gray',align='right')
#binThe edges of the bins.
#Length nbins + 1 (nbins left edges and right edge of last bin). 
#Always a single array even when multiple data sets are passed in.

ax[1].set_title('cumulative ')
ax[1].hist(n,color='orange',cumulative=True,bins=10,align="left")
```

```python
 #Draw contour map
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta=0.025
x=np.arange(-3.0,3.0,delta)
y=np.arange(-2.0,2.0,delta)
X,Y=np.meshgrid(x,y)#codinent 
zz=np.exp(-X**2 -Y**2)#Output array, element-wise exponential of x. This is a scalar if x is a scalar.


zzz=np.exp(-(X - 1)**2 - (Y - 1)**2)
Z=(zz - zzz) *2
print(X)
print(Y)

```

```python
fig,ax=plt.subplots()
CS=ax.contour(X,Y,Z)
ax.clabel(CS, inline=1, fontsize=10)
ax.set_title("Contour")
```

```python
#Draw 3D surface image
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig= plt.figure(figsize=(14,6))
#the 3d Grapic draw projection 3d
xn=fig.add_subplot(1,1,1, projection='3d')
xn.plot_surface(X,Y,Z,color='orange',rstride=4,cstride=4,linewidth=0)
```

# Practice Example
Write a Python programming to create a pie chart of the popularity of programming Language

```python
import matplotlib.pyplot as plt
#Data for the plot
language="python",'java','php','javascript','c#'
popurality=[22.2, 17.6, 8.8, 8.7, 7.2]
colors=["blue","green","red","skyblue","pink"]
explo=(0.1,0,0,0,0)#its use to explode the first slice
```

```python
#Explode the first slice

plt.pie(popurality, explode=explo, labels=language, colors=colors,autopct='%1.1f%%',shadow=True,startangle=150)
```
